```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow   # as colab doesn't allow cv2.imshow
```

```python
cam = cv2.VideoCapture("/content/number_plate_blur (2).avi")
```

```python
if not os.path.exists('data'):              # creating data directory to store our frames
    os.makedirs('data')
```

```python
currentframe = 0

while(True):

    # reading from frame
    ret,frame = cam.read()
    if ret:
    # if video is still left continue creating images
      name = './data/frame' + str(currentframe) + '.jpg'
      print (name)
 
      # writing the extracted images
      cv2.imwrite(name, frame)
  
      # increasing counter so that it will
      # show how many frames are created
      currentframe += 1
    else:
      break
```

```python
output_vid_dir = 'output video'
if not os.path.exists(output_vid_dir):
  os.mkdir(output_vid_dir)
```

```python
fps = 4
data_dir = 'data'
fourcc = cv2.VideoWriter_fourcc(*'mp4v') # used to specify the video codec
output_file_name = 'output_file.mp4'.format(data_dir, fps)
```

```python
# output video writer object 
out= cv2.VideoWriter(output_file_name, fourcc, fps, (h,w))
```

```python
def rotateimg(path):
  img = cv2.imread(path)
  num_rows, num_cols = img.shape[:2]
  rotation_matrix = cv2.getRotationMatrix2D((num_cols/2, num_rows/2), 180, 0.7)
  img_rotation = cv2.warpAffine(img, rotation_matrix, (num_cols, num_rows))
  out.write(img_rotation)      # write frame to video

  if img_rotation is not None:
    cv2_imshow(img_rotation)
    cv2.waitKey(1)
  
```

```python
for folder in os.listdir('/content/data'):
        rotateimg('/content/data'+'/'+folder)
out.release()
cv2.destroyAllWindows()
```
